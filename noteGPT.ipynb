````python
import config
import openai

import re
from IPython.core.getipython import get_ipython

import pandas as pd

import requests
from bs4 import BeautifulSoup

openai.api_key = config.api_key
````

````python
def set_chatgpt_cosplay():
    return input('dime que soy, por ejemplo:"Traductor", "Cientifico de Datos", "CEO de OpenAI"')
````

````python
def send_messanges_and_get_response(cosplay, messages):
    prompt = input("¿Que pregunta tienes para este "+cosplay+"?")
    messages.append({"role":"user", "content": prompt})
    response = openai.ChatCompletion.create(model="gpt-3.5-turbo", messages=messages)
    response_content = response.choices[0].message.content
    messages.append({"role":"assistant", "content": response_content})
    return response_content
````

````python
cosplay = set_chatgpt_cosplay()
````

````python
# Contexto de nuestro aplicación
context =  [{
    "role": "system",
    "content": "Eres un " + cosplay
}]

messages = context
````

````python
response = send_messanges_and_get_response(cosplay, messages)
print(response)
````

````python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Hacemos la petición a la página web
url = 'https://www.imdb.com/chart/top/?ref_=nv_mv_250'
response = requests.get(url)

# Parseamos el código HTML con BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Obtenemos las etiquetas que contienen la información que deseamos
movies_div = soup.find_all('td', class_='titleColumn')
ratings_div = soup.find_all('td', class_='ratingColumn imdbRating')

# Creamos los arrays donde almacenaremos los datos que obtengamos
titles = []
years = []
ratings = []

# Iteramos por cada película y extraemos su información
for movie in movies_div:
    title = movie.a.text
    year = movie.span.text.strip('()')
    titles.append(title)
    years.append(year)

# Iteramos por cada calificación y la almacenamos en el array de ratings
for rating in ratings_div:
    score = rating.strong.text
    ratings.append(score)

# Creamos el DataFrame
movies_df = pd.DataFrame({'title': titles, 'year': years, 'rating': ratings})

# Imprimimos el DataFrame
movies_df
````

````python
import matplotlib.pyplot as plt

# Creamos la gráfica con Matplotlib
plt.scatter(movies_df['year'], movies_df['rating'])

# Configuramos los labels del graph
plt.xlabel('Year')
plt.ylabel('Rating')
plt.title('IMDb Top 250')

# Mostramos la gráfica
plt.show()
````

````python
def add_code_cell(cell_text):
    shell = get_ipython()
    payload = dict(
        source='set_next_input',
        text=cell_text,
        replace=False,
    )
    shell.payload_manager.write_payload(payload, single=False)
````

````python
def get_and_add_python_code_from_gpt():
    response = send_messanges_and_get_response(cosplay, messages)

    pattern = r"```python\n([\s\S]*?)\n```"

    matches = re.findall(pattern, response)

    if matches:
        for match in matches:
            code = match
            add_code_cell(code)
````

````python
get_and_add_python_code_from_gpt()
````
